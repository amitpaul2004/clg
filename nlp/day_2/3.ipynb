```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
S = "The quick brown fox jumps over the lazy dog."
```

```python
tokens = word_tokenize(S)
tagged = pos_tag(tokens)
```

```python
nouns_count = len([word for word, tag in tagged if tag.startswith('NN')])
verbs_count = len([word for word, tag in tagged if tag.startswith('VB')])
```

```python
print(f"{nouns_count} {verbs_count}")
```
