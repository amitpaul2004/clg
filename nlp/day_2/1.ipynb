```python
# Cell 1: Define the input string
S = "Python is a powerful language for Data Science"
# Alternative: S = input("Enter your paragraph: ")
```

```python
# Cell 2: Break the string into a list of words
tokens = S.split()

# Let's see what the list looks like
print(f"List of tokens: {tokens}")
```

```python
# Cell 3: Final Output
# 1. Print the total count
print(len(tokens))

# 2. Print each token in lowercase
for word in tokens:
    print(word.lower())
```
