```python
import nltk
from nltk import CFG
```

```python
grammar = CFG.fromstring("""
  S -> NP VP
  NP -> Det N
  VP -> V NP
  Det -> 'the' | 'a'
  N -> 'fox' | 'dog'
  V -> 'chases'
""")
```

```python
sentence = "the fox chases a dog".split()
```

```python
parser = nltk.ChartParser(grammar)
```

```python
for tree in parser.parse(sentence):
    print("Bracketed Structure:")
    print(tree)
    print("\nVisual Parse Tree:")
    # Ensure this line has exactly one indentation (4 spaces)
    tree.pretty_print()
```
