# Logistic Regression - Insurance Prediction

This notebook analyzes `insurance_data.csv` to predict whether a customer will buy insurance based on their age.

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
%matplotlib inline
```

### Load the dataset

```python
df = pd.read_csv('../data/insurance_data.csv')
df.head()
```

### Visualize the data

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Age vs Insurance Purchase')
```

### Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.8, random_state=42)
X_test
```

### Create and train the model

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

### Inspect Model Parameters
Here we'll check the coefficient (weight) and the intercept (bias) of our logistic regression model.

```python
print("Coefficient (m):", model.coef_[0][0])
print("Intercept (b):", model.intercept_[0])
```

### Logistic Regression 'Slope' (Sigmoid Curve)

The predicted probabilities follow a sigmoid curve $y = \frac{1}{1 + e^{-(mx+b)}}$, where $m$ is the slope (coefficient) and $b$ is the intercept.

```python
# Create a range of ages for the curve
X_range = np.linspace(df.age.min(), df.age.max(), 100).reshape(-1, 1)
y_prob = model.predict_proba(X_range)[:, 1]

plt.scatter(df.age, df.bought_insurance, marker='+', color='red', label='Data')
plt.plot(X_range, y_prob, color='blue', label='Sigmoid Curve (Slope)')
plt.xlabel('Age')
plt.ylabel('Bought Insurance Probability')
plt.title('Visualizing the Logistic Regression Slope')
plt.legend()
plt.show()
```

### Model Predictions and Probabilities

The table below shows the age, the model's final prediction, and the internal probabilities for each class (0 = won't buy, 1 = will buy).

```python
results = pd.DataFrame(X_test, columns=['age'])
results['predicted'] = model.predict(X_test)
prob = model.predict_proba(X_test)
results['prob_0'] = prob[:, 0]
results['prob_1'] = prob[:, 1]
results
```

### Model Performance

```python
accuracy = model.score(X_test, y_test)
print(f"Model Accuracy: {accuracy * 100}%")
```
